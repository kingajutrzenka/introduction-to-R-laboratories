### 1. (3p) Utwórz dwa wektory liczbowe: 
a) składający się z ciągu liczb całkowitych od -10 do 5 

b) składający się z ciągu liczb rozpoczynających się od 6 z krokiem 2 długości 16.

c) połącz te dwa wektory kolumnowo.   

```r
#a

w1 <- c(-10:5)
w1
```

```r
#b

#help(seq)

w2 <- seq(from=6, by=2, length = 16)
w2
```

```r
#c

w3 <- cbind(w1, w2)
w3

#cbind - łączenie danych kolumnowo
#rbind - łączenie danych wierszowo
```
